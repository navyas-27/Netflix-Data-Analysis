```python
#Find latest movies
top_movies = df.sort_values(by='Release_Date',ascending=False)

top_movies[['Title','Release_Date']].head(10)
```

```python
#Search a movie
movie=input("Enter movie:")
result=df[df["Title"].str.contains(movie,case=False)]
print(result[["Title","Type"]])
```

```python
df["Country"].value_counts().head(20)
```

```python
import matplotlib.pyplot as plt
c=(df["Country"].str.split(",").explode()
   .str.strip().value_counts())
plt.figure(figsize=(8,6))
c.head(15).plot(kind="bar")
plt.title("Top 10 countries on Netflix")
plt.xlabel("Country")
plt.ylabel("No. of movies")
plt.show()
```

```python
# Rating Distribution
rating=df["Rating"].value_counts()
rating.plot.pie()
plt.title("Content rating")
plt.show()
```

```python
import numpy as np
```

```python
movies=df[df["Category"]=="Movie"].copy()
movies["Minutes"]=movies["Duration"].str.replace("min"," ").astype(int)
duration=movies["Minutes"].to_numpy()
print(duration)
```

```python
print("Average Duration:",np.mean(duration))
print("Longest Movie :", np.max(duration))
print("Shortest Movie :", np.min(duration))
print("Median Duration :", np.median(duration))
```
